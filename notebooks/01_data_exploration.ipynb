```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path
import warnings
warnings.filterwarnings('ignore')

# Set display options
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 100)
plt.style.use('default')
sns.set_palette("husl")
```

```python
# Load the Excel file
excel_file_path = Path('../data/raw/Bellary.xlsx')
print(f"Loading Excel file: {excel_file_path}")
print(f"File exists: {excel_file_path.exists()}")
print(f"File size: {excel_file_path.stat().st_size / (1024*1024):.2f} MB")
```

```python
# Read Excel file and get sheet names
xls = pd.ExcelFile(excel_file_path)
sheet_names = xls.sheet_names

print(f"Number of sheets: {len(sheet_names)}")
print("Sheet names:")
for i, sheet in enumerate(sheet_names, 1):
    print(f"  {i}. {sheet}")
```

```python
# Check columns in each sheet
sheet_columns = {}
for sheet_name in sheet_names:
    df = pd.read_excel(xls, sheet_name=sheet_name)
    sheet_columns[sheet_name] = list(df.columns)
    print(f"\n{sheet_name} ({df.shape[0]} rows, {df.shape[1]} columns):")
    
    # print("Columns:", df.columns.tolist())
```

```python
# Let's examine the first few rows of each sheet to understand the data structure


for sheet_name in sheet_names:
    print(f"\n{'='*60}")
    print(f"SHEET: {sheet_name}")
    print(f"{'='*60}")
    
    df = pd.read_excel(xls, sheet_name=sheet_name)
    
    print(f"Shape: {df.shape[0]} rows × {df.shape[1]} columns")
    print(f"\nColumn names:")
    for i, col in enumerate(df.columns, 1):
        print(f"  {i:2d}. {col}")
    
    print(f"\nFirst 10 rows:")
    display(df.head(10))
    
    # print(f"\nData types:")
    # for col in df.columns:
    #     print(f"  {col}: {df[col].dtype}")
    
    # # Break after first sheet for now to see the structure
    # break
```

```python
# Function to clean individual sheet data
def clean_sheet_data(df, sheet_name):
    """
    Clean sheet data by removing status-only rows and adding metadata
    """
    print(f"Cleaning {sheet_name}...")
    print(f"  Original rows: {len(df)}")
    
    # Create a copy to avoid modifying original
    df_clean = df.copy()
    
    # Identify rows that are meaningful (have farmer data)
    # These rows should have at least Farmer Code filled
    meaningful_rows = df_clean['Farmer Code'].notna()
    
    # Filter to keep only meaningful rows
    df_clean = df_clean[meaningful_rows].copy()
    
    # Add source sheet information
    df_clean['source_sheet'] = sheet_name
    
    # Reset index
    df_clean = df_clean.reset_index(drop=True)
    
    print(f"  Cleaned rows: {len(df_clean)}")
    print(f"  Removed: {len(df) - len(df_clean)} status-only rows")
    
    return df_clean

# Process all sheets
cleaned_sheets = []
total_original = 0
total_cleaned = 0

for sheet_name in sheet_names:
    df = pd.read_excel(xls, sheet_name=sheet_name)
    total_original += len(df)
    
    df_clean = clean_sheet_data(df, sheet_name)
    cleaned_sheets.append(df_clean)
    total_cleaned += len(df_clean)

print(f"\n{'='*50}")
print(f"CLEANING SUMMARY")
print(f"{'='*50}")
print(f"Total original rows: {total_original:,}")
print(f"Total cleaned rows: {total_cleaned:,}")
print(f"Rows removed: {total_original - total_cleaned:,}")
print(f"Data retention: {(total_cleaned/total_original)*100:.1f}%")
```

```python
# Combine all cleaned sheets into a single DataFrame
print("Combining all sheets...")
master_df = pd.concat(cleaned_sheets, ignore_index=True)

print(f"\nMASTER DATASET CREATED")
print(f"{'='*30}")
print(f"Total records: {len(master_df):,}")
print(f"Total columns: {len(master_df.columns)}")
print(f"Memory usage: {master_df.memory_usage(deep=True).sum() / 1024:.2f} KB")

# Show breakdown by source sheet
print(f"\nRecords by source sheet:")
sheet_counts = master_df['source_sheet'].value_counts()
for sheet, count in sheet_counts.items():
    print(f"  {sheet}: {count:,} records")

# Show sample of combined data
print(f"\nSample of combined data:")
display(master_df.head())
```

```python
# Final data quality check before export
print("FINAL DATA QUALITY CHECK")
print("="*40)

# Check for duplicates
duplicates = master_df.duplicated(subset=['Farmer Code']).sum()
print(f"Duplicate Farmer Codes: {duplicates}")

if duplicates > 0:
    print("Duplicate farmer codes found:")
    duplicate_codes = master_df[master_df.duplicated(subset=['Farmer Code'], keep=False)]['Farmer Code'].unique()
    for code in duplicate_codes[:5]:  # Show first 5
        print(f"  {code}")

# Check missing values in key fields
key_fields = ['Farmer Code', 'Farmer Name', 'Village']
print(f"\nMissing values in key fields:")
for field in key_fields:
    missing = master_df[field].isna().sum()
    print(f"  {field}: {missing} ({missing/len(master_df)*100:.1f}%)")

# Check data types
print(f"\nData types:")
for col in master_df.columns:
    print(f"  {col}: {master_df[col].dtype}")

# Show unique values for categorical fields
categorical_fields = ['Status']
print(f"\nUnique values in categorical fields:")
for field in categorical_fields:
    unique_vals = master_df[field].value_counts()
    print(f"  {field}: {unique_vals.to_dict()}")
```

```python
# Export to CSV
from datetime import datetime

# Create output filename with timestamp
timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
csv_filename = f"bellary_farmers_cleaned_{timestamp}.csv"
csv_path = Path("../data/processed") / csv_filename

# Ensure processed directory exists
csv_path.parent.mkdir(exist_ok=True)

# Export to CSV
print(f"Exporting to CSV...")
print(f"File: {csv_path}")

master_df.to_csv(csv_path, index=False)

# Verify the export
if csv_path.exists():
    file_size = csv_path.stat().st_size / 1024  # KB
    print(f"✅ Export successful!")
    print(f"   File size: {file_size:.2f} KB")
    print(f"   Records exported: {len(master_df):,}")
    
    # Also create a standard filename for easy access
    standard_path = Path("../data/processed") / "bellary_farmers_clean.csv"
    master_df.to_csv(standard_path, index=False)
    print(f"   Also saved as: {standard_path}")
else:
    print(f"❌ Export failed!")

print(f"\n🎉 DATA CONVERSION COMPLETE!")
print(f"Next step: Load this CSV into PostgreSQL database")
```
